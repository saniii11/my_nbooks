```python
import pandas as pd
import seaborn as sns
```

```python
avocado_url = 'avocado.csv'
```

```python
avocado = pd.read_csv(avocado_url)
```

```python
avocado.head()
```

```python
avocado.info()
```

```python
avocado['Date'].apply(pd.to_datetime)
```

```python
avocado.info()
```

```python
avocado.query('year == 2015')[['Date', 'Total Bags', 'Small Bags']].head()
```

```python
avocado_melted = pd.melt(avocado, id_vars='Date', value_vars=['Total Bags', 'Small Bags'],
                    var_name='Bags', value_name='Count')
avocado_melted
```

```python
sns.relplot(data=avocado_melted, kind='line', x='Date', y='Count', hue='Bags')
```

```python
avocado_filtered = avocado.copy()
avocado_filtered['Sales Volume'] = pd.qcut(avocado['Total Volume'], q=4, labels=['Poor','Modest','Good','Excellent'])
```

```python
avocado_filtered.head()
```

```python
sns.countplot(data=avocado_filtered, x='year', hue='Sales Volume')
sales_volume_data = avocado_filtered.groupby(['year', 'Sales Volume']).size().reset_index(name='Count')
```
