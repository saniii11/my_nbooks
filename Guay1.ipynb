## Get the data

```python
import pandas as pd
poll_url = 'https://www.murach.com/python_analysis/president_polls_2016.csv'
polls = pd.read_csv(poll_url)
polls.head()
```

## Sort the data

```python
polls.sort_values('state', ascending=False).head()
```

## Use Tab completion and tooltips

```python
polls.sort_values('state', ascending=False).head()
```

## Use lists, slices, tuples, and dictionary objects

### A list

```python
polls.columns
```

```python
polls.drop(columns=['cycle','branch','matchup','forecastdate']).head(2)
```

### A tuple

```python
polls.enddate.unique
```

```python
polls.query("enddate >= '2016-06-01'")
```

```python
polls['enddate'] = pd.to_datetime(polls.enddate, format='%m/%d/%Y')
polls.plot.line(x='enddate', y="adjpoll_clinton", xlim=('9/22/2016', '10/1/2016'))
```

### A dictionary

```python
polls.rename(columns={'adjpoll_clinton':'Clinton',
                      'adjpoll_trump':'Trump'})
```

### Two slices

```python
polls.loc[0:100:10,'state':'grade']
```
